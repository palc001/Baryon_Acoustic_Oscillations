```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
arr1 = np.array(20.*np.random.random_sample((1000)) - 10.)

arr2, remove = [], []

for i in range(len(arr1)):
    arr2.extend((arr1[i]-1., arr1[i]+1.))
for i in range(len(arr2)):
    if (arr2[i] < -10. or arr2[i] > 10.): remove.append(i)
arr3 = np.delete(arr2, remove)
arr3.sort()

binsize = 0.1
bins = np.arange(-1, 1+binsize, binsize)
totcounter = 0
totcounts = np.zeros(len(bins)-1)

for i in range(len(arr3)):
    if (arr3[i] < -9. or arr3[i] > 9.): continue
    temp = []
    for j in range(len(arr3)):
        if (arr3[i] < -9. or arr3[i] > 9. or i == j): continue
        dist = arr3[j]-arr3[i]
        if (-1. < dist < 1.): temp.append(dist)
    tempcounts, binedges = np.histogram(temp, bins)
    totcounter += 1
    for k in range(len(totcounts)): totcounts[k] += tempcounts[k] 

avgcounts = totcounts/totcounter
```

```python
center = (bins[:-1] + bins[1:]) / 2.
plt.plot(center, avgcounts)
plt.show()
```
